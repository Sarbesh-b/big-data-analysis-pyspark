```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the DataFrame 
df = pd.read_csv("Task 3 Q7.csv")

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
axs = axs.flatten()  

# 1) HISTOGRAM: FARE DISTRIBUTION
axs[0].hist(df["fare_amount"], bins=50, range=(0, 100), color="pink", edgecolor="black")
axs[0].set_title("Fare Distribution")
axs[0].set_xlabel("Fare Amount ($)")
axs[0].set_ylabel("Frequency")

# 2) HISTOGRAM: TRIP DISTANCE
axs[1].hist(df["distance"], bins=50, range=(0, 100), color="blue", edgecolor="black")
axs[1].set_title("Trip Distance Distribution")
axs[1].set_xlabel("Distance (miles)")
axs[1].set_ylabel("Frequency")

# 3) HISTOGRAM: TRIP DURATION
axs[2].hist(df["duration"], bins=50, range=(0, 100), color="red", edgecolor="black")
axs[2].set_title("Trip Duration Distribution")
axs[2].set_xlabel("Duration (seconds)")
axs[2].set_ylabel("Frequency")

# 4) HISTOGRAM: START HOUR
axs[3].hist(df["start_hour"], bins=24, range=(0, 24), color="purple", edgecolor="black")
axs[3].set_title("Distribution by Start Hour")
axs[3].set_xlabel("Hour of Day (0-23)")
axs[3].set_ylabel("Frequency")

plt.tight_layout()  # Adjust layout so subplots don't overlap
plt.show()
```
