Question 3

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the merged CSV file for Question 3, which contains the rating distribution.
df = pd.read_csv("Task2_Q3.csv")

# Preview the data: should have columns like 'rating_category' and 'count'
print("Rating distribution data:")
print(df.head())

# Convert the 'count' column to numeric (if needed)
df['count'] = pd.to_numeric(df['count'], errors='coerce')

# --- Create a Pie Chart ---
# We use the rating_category as labels and the count as values.
labels = df['rating_category']
sizes = df['count']

# Optionally, define an "explode" setting to pull out each slice slightly
explode = [0.05] * len(labels)  # Slightly "explode" each slice

# Define custom colors for each category (adjust the colors as desired)
colors = ['red', 'purple', 'lightskyblue']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, explode=explode, shadow=True, colors=colors)
plt.title("Rating Distribution by Category")
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.
plt.show()
```

Question 4

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the new CSV file for Q4 aggregated output.
# (Change the filename if necessary)
df = pd.read_csv("task2_q4_time_of_year_distribution.csv")

# Preview the data to ensure it looks correct.
print("Q4: Time distribution data:")
print(df.head())

# Convert the "time_of_year" column to an ordered categorical variable.
df['time_of_year'] = pd.Categorical(df['time_of_year'], categories=["Early Year", "Late Year"], ordered=True)
df = df.sort_values("time_of_year")

# Create a bar chart.
plt.figure(figsize=(8, 5))
bars = plt.bar(df['time_of_year'], df['count'], color='red', edgecolor='black')

# Add labels and title.
plt.xlabel("Time of Year")
plt.ylabel("Number of Ratings")
plt.title("Number of Ratings by Time of Year")
plt.ylim(1.65e7, 1.71e7)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Annotate each bar with the corresponding count.
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{int(height)}',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Display the plot.
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the aggregated CSV output for Q5 Genre Distribution.
# Ensure that the file "task2_q5_genre_distribution.csv" (merged CSV) is in your current directory.
df = pd.read_csv("task2_q5_genre_distribution.csv")

# Preview the data to make sure it looks as expected.
print("Genre Distribution Data:")
print(df.head())

# Sort the DataFrame by genre (optionally, you could define a custom order if needed)
df = df.sort_values("genre")

# Create a bar chart.
plt.figure(figsize=(10, 6))
plt.bar(df["genre"], df["count"], color="brown")
plt.xlabel("Genre")
plt.ylabel("Number of Ratings")
plt.title("Genre Distribution: Number of Ratings per Genre")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("task2_q6_temporal_analysis.csv")

# Convert columns to proper numeric types
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['count'] = pd.to_numeric(df['count'], errors='coerce')

# Sort by year so the plot goes in ascending year order
df = df.sort_values("year")

# Create figure
plt.figure(figsize=(10, 6))

# Plot line chart
plt.plot(df['year'], df['count'], marker='o', linestyle='-', color='b')

# Label the axes and set the title
plt.xlabel("Year")
plt.ylabel("Total Number of Ratings")
plt.title("Temporal Analysis: Total Number of Ratings per Year")

# Generate a sorted list of distinct integer years
years = df['year'].dropna().unique().astype(int)
years.sort()

# Option 1: Show every year, but rotate labels
plt.xticks(years, rotation=45, ha="right")


plt.grid(True)
plt.tight_layout()  # Ensure labels fit
plt.show()
```
